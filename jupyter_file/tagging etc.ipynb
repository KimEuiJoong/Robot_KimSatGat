```python
#태그단거 합치기
import pandas as pd
import pickle

poem_tag1 = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\\tagging\\1028\poems_tag_1028.csv')
poem_tag2 = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\\tagging\\1028\poems_tag_v1028_KHS.csv')
poem_tag3 = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\\tagging\\1028\poems_tag_v1025_KEJ.csv')
poem_tag4 = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\\tagging\\1028\poems_tag_RJH_v1028.csv')
poem_tag1 = poem_tag1[poem_tag1!='#']
poem_tag1.dropna(axis=0, inplace=True)
poem_tag2 = poem_tag2[poem_tag2!='#']
poem_tag2.dropna(axis=0, inplace=True)
poem_tag3 = poem_tag3[poem_tag3!='#']
poem_tag3.dropna(axis=0, inplace=True)
poem_tag4 = poem_tag4[poem_tag4!='#']
poem_tag4.dropna(axis=0, inplace=True)

#4개 합치고 중복 제거
poem_tag = pd.concat([poem_tag1['tags'], poem_tag2['tags'], poem_tag3['tags'], poem_tag4['tags']])
poem_tag = poem_tag.groupby(level=0).first()

poem_tag
```

```python
#..
poems = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\poems.csv', names=['name','writer','text','cc'])

poems['tags'] = poem_tag
poems.fillna('#', inplace=True)

print(len(poems[poems['tags']!='#']))

poems
```
