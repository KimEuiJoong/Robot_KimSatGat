```python
#1.1
#시 본문 데이터 불러오기
import pandas as pd
import numpy as np

poems_df = pd.read_csv('C:\DeepLearing\Robot_KSG\poem\poems_preprocessing_v1.csv')
poems_df = poems_df['text']

poems_df[:3]
```

```python
#1.2
#리스트로 변환
poems = []

for i in range(len(poems_df)-1):
    poems.append(poems_df[i])

poems[:2]
```

```python
#2.1
#Customized KoNLPy 사용 준비
from ckonlpy.tag import Twitter
from ckonlpy.tag import Postprocessor
from ckonlpy.utils import load_wordset

twitter = Twitter(use_twitter_dictionary=False) #default=True

#stopwords = load_wordset('C:\DeepLearing\Robot_KSG\poem\stopwords.txt')
passtags = ['Noun', 'Verb', 'Adjective', 'Adverb', 'Modifier', 'Exclamation', 'Unknown'] 
#Modifier: 관형사, Exclamation: 감탄사

postprocessor = Postprocessor(
    base_tagger = twitter,
    #stopwords = stopwords,
    passtags = passtags
)
```

```python
#2.2
#Customized KoNLPy 이용해 토큰화
poems_token = []

for poem in poems:
    poems_token.append(postprocessor.pos(poem))

poems_token[:3]
```

```python
#2.3
#품사 태그 제거
poems_tokens = []

for poem in poems_token:
    poem_tokens = []
    for word, tag in poem:
        poem_tokens.append(word)
    poems_tokens.append(poem_tokens)

poems_tokens[:2]
```

```python
temp_dic = {'column1': poems_tokens}
poems_tokens_df = pd.DataFrame(temp_dic)

poems_tokens_df
```

```python
poems_tokens_df = poems_tokens_df['column1']
poems_tokens_df
```

```python
#3.1
#1차 정수 인코딩
from gensim import corpora
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer()
tokenizer.fit_on_texts(poems_tokens_df)

print(tokenizer.word_index)
```

```python
#3.2
#단어 비율 확인

threshold = 3 #threshold: 임계값
total_cnt = len(tokenizer.word_index) #total_cnt: 전체 단어 수
rare_cnt = 0 #rare_cnt: 등장 빈도수 0~2인 단어 count
total_freq = 0 #total_freq: 훈련 데이터의 전체 단어 빈도수 총 합
rare_freq = 0 #rare_freq: 등장 빈도수 0~2인 단어의 빈도수 총 합

for key, value in tokenizer.word_counts.items(): #key, value: 단어, 빈도수
    total_freq += value
    
    if(value < threshold):
        rare_cnt += 1
        rare_freq += value

print('vocabulary 크기 : ', total_cnt)
print('등장 빈도 %s 이하인 희귀 단어 수 : %s' %(threshold-1, rare_cnt))
print("희귀 단어 비율 : ", (rare_cnt/total_cnt)*100)
print("전체 등장 빈도에서 희귀 단어 등장 빈도 비율 : ", (rare_freq/total_freq)*100)
```

```python
#3.3
#빈도수 3 미만 단어 제거하며 정수 인코딩 + 정수-단어 사전 생성

vocab_size = total_cnt - rare_cnt + 2 #+2: 0번- 패딩토큰, 1번- OOV토큰

tokenizer = Tokenizer(vocab_size, oov_token = 'OOV') #vocab_size 이상의 정수는 OOV로 변환
tokenizer.fit_on_texts(poems_tokens_df)

dictionary = tokenizer.word_index 
dictionary = {v:k for k, v in dictionary.items() if v<=vocab_size} #key, value: 정수, 단어 로 수정(vocab_size만큼만 가져와서 뒤집기)

poems_tokens_int = tokenizer.texts_to_sequences(poems_tokens_df)

print(len(dictionary))
poems_tokens_int[1]
```

```python
#빈도수 낮은 데이터로만 구성된 샘플 있나 확인
drop_train = [index for index, sentence in enumerate(poems_tokens_int) if len(sentence)<1]

print(drop_train)
```

```python
#4.0
#LDA를 위한 전처리 함수 생성: 정수 토큰 -> (정수, 빈도수)로 변환
def corpus_int_frq(corpus):
    corpus_return = []
    corpus_temp = []
    
    for int in corpus:
        if(int!=1): #OOV는 추가 안함(OOV아닐시 추가)
            corpus_temp.append((int, corpus.count(int)))
    
    for data in corpus_temp:
        if data not in corpus_return:
            corpus_return.append(data)
    
    return corpus_return
```

```python
cp = corpus_int_frq(poems_tokens_int[1])

cp[:15]
```

```python
#4.1
#LDA를 위한 전처리
poems_token_LDA_ckonlpy = []
for poem in poems_tokens_int:
    poems_token_LDA_ckonlpy.append(corpus_int_frq(poem))

poems_token_LDA_ckonlpy[1][:10]
```

```python
#4.2
#(정수, 빈도수)토큰화 된 시들 내보내기
import pickle

with open('C:\DeepLearing\Robot_KSG\LDA\corpus1_int_frq', 'wb') as fw:
    pickle.dump(poems_token_LDA_ckonlpy, fw)

with open('C:\DeepLearing\Robot_KSG\LDA\dictionary1', 'wb') as fw:
    pickle.dump(dictionary, fw)
```

```python
#5.1
#다중 클래스 분류를 위한 전처리(불용어 제거한 토큰 상태로 rollback)
poems_token_MLP = []
for poem in poems_tokens_int:
    poem_tokens = []
    for iword in poem:
        if iword != 1:
            poem_tokens.append(dictionary[iword])
    poems_token_MLP.append(poem_tokens)

poems_token_MLP[:2]
```
