```python
#1
#시 token data 불러오기
import pandas as pd
import numpy as np

poems_token = pd.read_csv("C:\DeepLearing\Robot_KSG\poem\poems_text_token.csv", index_col=0) #index_col=0: 맨앞의 열 인덱스로 사용(인덱스 재부여 방지)
poems_token.fillna('-', inplace=True) #NaN -> '-' 로 변환(제거를 위해)

poems_token[:3]
```

```python
#2.1
#리스트로 변환
poems_token = poems_token.values.tolist()
```

```python
#2.2
#NaN 제거
for poem_token in poems_token:
    while '-' in poem_token:
        poem_token.remove('-')

poems_token[10]
```

```python
#3
#정수 인코딩
from gensim import corpora

dictionary = corpora.Dictionary(poems_token) #단어에 정수 인코딩 + 사전에 저장
corpus = [dictionary.doc2bow(text) for text in poems_token] #시 단어 토큰->정수, 빈도수 로 변환

print(corpus[1])
```

```python
dictionary[0]
```

```python
len(dictionary)
```

```python
#4
#LDA 모델 훈련
import gensim

NUM_TOPICS = 20
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=50)
#corpus: 정수 인코딩된 단어 토큰 말뭉치, passes: 알고리즘 동작 횟수
topics = ldamodel.print_topics(num_words=10) #num_words: topic당 나타낼 단어 수

for topic in topics:
    print(topic)
```

```python
#4'
#LDA 모델 훈련
import gensim

NUM_TOPICS = 50
ldamodel2 = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=100)
#corpus: 정수 인코딩된 단어 토큰 말뭉치, passes: 알고리즘 동작 횟수
topics2 = ldamodel2.print_topics(num_words=15) #num_words: topic당 나타낼 단어 수

for topic in topics2:
    print(topic)
```

```python
#5
#LDA 모델 시각화
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
#람다: 1-토픽 별로 자주 등장하는 단어들을 키워드로 선택 / 0-토픽 간에 차이가 많이 나는 단어를 선택
```

```python
#5'
#LDA 모델 시각화

vis2 = pyLDAvis.gensim.prepare(ldamodel2, corpus, dictionary)
pyLDAvis.display(vis2)
```

```python
#6
#html파일로 출력
pyLDAvis.save_html(vis2, 'topic50_word15_passes100.html')
```
