```python
#1
#시 row data -> DataFrame으로 변환
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_data = pd.read_csv("C:\DeepLearing\Robot_KSG\poem\poems.csv", names=['제목','시인','본문','etc']) #시 데이터 DataFrame 형태로 불러오기
csv_data = csv_data.drop(['etc'], axis='columns') #etc 열 삭제

csv_data
```

```python
#2.1
#불용어 처리1 - 한글,숫자 빼고 제거

csv_data['제목'] = csv_data['제목'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]", "")
csv_data['본문'] = csv_data['본문'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", "")
```

```python
#2.2
#mecab으로 data token화
from eunjeon import Mecab

mecab = Mecab()
poems_token = []

for i in range(len(csv_data)): #i: 시 index
    for j in range(3): #j: column index
        poems_token.append(mecab.morphs(csv_data.loc[i][j])) #토큰화 후 poems_token에 1차원 배열로 저장
```

```python
#2.3
#DataFrame으로 변환
poems_token = np.array(poems_token).reshape(-1,3) #numpy array로 변환
poems_token = pd.DataFrame(poems_token, columns=['제목','시인','본문']) #DataFrame으로 변환

poems_token
```

```python
#2.4
#불용어 처리2 - token 중 불용어 제거

stopwords = np.loadtxt("C:\DeepLearing\Robot_KSG\poem\stopwords.csv", delimiter=',', dtype='str') #stopwords 불러오기

poems_token['제목'] = poems_token['제목'].apply(lambda x: [item for item in x if item not in stopwords])
poems_token['본문'] = poems_token['본문'].apply(lambda x: [item for item in x if item not in stopwords])

poems_token
```

```python
#2.5.1
#csv파일로 내보내기
#poems_token_mecab.csv: 제목, 시인, 본문 토큰화 되어있는 csv 파일

poems_token.to_csv("C:\DeepLearing\Robot_KSG\poem\poems_token_mecab.csv", sep=',', na_rep='NaN')
```

```python
#2.5.2
#본문 데이터만 정제 및 csv,txt파일로 내보내기
#poems_text_token.csv: 시 본문 토큰화 상태/ poems.txt: 시 본문 데이터

poems_text_token = poems_token['본문'].apply(lambda x: pd.Series(x)) #본문만 다시 dataframe화
poems_text_token.dropna(axis=1, thresh=487, inplace=True) #데이터 10% 이하(NaN 90%이상)인 열 제거 / thresh: 결측치 수치 지정, inplace: 현재 dataframe에 바로 적용

poems_text_token.to_csv("C:\DeepLearing\Robot_KSG\poem\poems_text_token.csv", sep=',', na_rep='NaN')

csv_data['본문'].to_csv('C:\DeepLearing\Robot_KSG\poem\poems.txt', index=False, header=None, sep=' ')
```
