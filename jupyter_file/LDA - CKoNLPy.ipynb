```python
#1
#시 토큰 데이터 및 단어 사전 불러오기 - (정수, 빈도수)
import pandas as pd
import numpy as np
import pickle
from gensim import corpora
from gensim.corpora import Dictionary

with open('C:\DeepLearing\Robot_KSG\LDA\corpus1_int_frq', 'rb') as fr:
    poems_token = pickle.load(fr)

with open('C:\DeepLearing\Robot_KSG\LDA\dictionary1', 'rb') as fr:
    dictionary = pickle.load(fr)

#gensim.corpora.Dictionary 객체로 변환(메소드 이용하기 위해)(dct->corpus->dct)
dct = []
for k, v in dictionary.items():
    dct.append([v])
dictionary = corpora.Dictionary(dct)

print(poems_token[:2])
print(dictionary)
```

```python
#불용어 제거됐는지 확인
a = 0
for integer, word in dictionary.items():
    if word == '사랑':
        a = integer
print(a)
```

```python
#2
#LDA 모델 훈련
import gensim

NUM_TOPICS = 50
ldamodel = gensim.models.ldamodel.LdaModel(
    poems_token, 
    num_topics = NUM_TOPICS, 
    id2word=dictionary,
    passes = 100
)
topics = ldamodel.print_topics(num_words=10)

for topic in topics:
    print(topic)
```

```python
#3
#LDA 모델 시각화
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, poems_token, dictionary)
pyLDAvis.display(vis)
```
